```python
#dataframe that has all the customers who have churned
churneddf = churndb.where(churndb.has_churned == True)
churneddf = churneddf.dropna()
```

```python
churneddf.groupby(['contract_description', 'partner_dependents']).partner_dependents.count()
```

```python
#dataframe that has all the month to month customers
mtmdf = churndb.where(churndb.contract_type == 0)
mtmdf = mtmdf.dropna()
```

```python
#0	for no partners and no dependents
#1	indicates partner only
#2	indicates dependents only 
#3	indicates partner & dependents
# create data frame for the dependents key

familykey = [(0, 'no partner no dependents'),
            (1, 'partner only'),
            (2, 'dependents only'),
            (3, 'partner & dependents')]
familykeydf = pd.DataFrame(familykey, 
                          columns = ['fam_id', 'fam_description'])
```

```python
familykeydf
```

```python
# merge with family key to get descriptions of family in dataframe
mtmdf = mtmdf.merge(familykeydf, left_on= 'partner_dependents', right_on = 'fam_id', how = 'right', indicator = False)
mtmdf.head()
```

```python
#do the same with the churned df
churneddf = churneddf.merge(familykeydf, left_on= 'partner_dependents', right_on = 'fam_id', how = 'right', indicator = False)
churneddf.head()
```

### Now have two dataframes churneddf (all the info for people who churned), and mtmdf (all the info for people who are month to month)

People on month to month who churn are paying more per month
- customers who churned pay on average 18.81 percent higher on their monthly bill
- that's 11.56 more per month
- that's also above the entire population average which is 64.76

To calculate percent higher
- ((N2-N1)/N1)*100 = PH
- ((73.02-61.46)/61.46)*100=18.81%
